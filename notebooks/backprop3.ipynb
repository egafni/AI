```python
import numpy as np

def relu(x):
    return np.where(x>0, x, 0)
    
def softmax(x):
    x = x - x.max(1, keepdims=True)
    x_exp = np.exp(x)
    return x_exp/x_exp.sum(1,keepdims=True)

B=10
H=100
C=20

X = np.random.randn(B,C)
y = np.random.randint(0,4,size=B)

w1 = np.random.randn(C, H) / np.sqrt(C) / np.sqrt(2)
b1 = np.zeros(H)
w2 = np.random.randn(H,H) / np.sqrt(H) / np.sqrt(2)
b2 = np.zeros(H) 
proj_w = np.random.randn(H, 4) / np.sqrt(H) / np.sqrt(2)
proj_b = np.zeros(4)

params = [w1,b1,w2,b2,proj_w,proj_b]

for i in range(20):
    losses = []
    # for h in [0,.]:
    # forward
    # b1[1]+=h
    pre1 = X@w1+b1
    act1 = relu(pre1)
    pre2 = act1@w2+b2
    act2 = relu(pre2)
    logits = act2@proj_w + proj_b

    probas = softmax(logits)
    log_probs = np.log(probas[range(B), y])
    
    loss = -log_probs.mean(0)
    losses.append(loss)
    print(f'loss: {loss}')
    # print('d/dx',(losses[1] - losses[0]) / h)

    dlogits = probas.copy()
    dlogits[range(B), y] -= 1
    dlogits /= B
    dact2 = dlogits@proj_w.T
    dproj_b = dlogits.sum(0)
    dproj_w = act2.T@dlogits
    dpre2 = np.where(pre2>0, dact2, 0.0)
    dact1 = dpre2@w2.T
    dw2 = act1.T@dpre2
    db2 = dpre2.sum(0)
    dpre1 = np.where(pre1>0,dact1,0.0)
    dw1 = X.T@dpre1
    db1 = dpre1.sum(0)

    dparams = [dw1,db1,dw2,db2,dproj_w,dproj_b]

    for dparam,param in zip(dparams,params):
        assert dparam.shape == param.shape
        lr = 1e-1
        param -= lr * dparam
```
